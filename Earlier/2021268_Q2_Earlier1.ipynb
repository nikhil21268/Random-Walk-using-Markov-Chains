```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
p = 0.5
num_steps = 500

# Simulate the random walk
positions = np.cumsum(np.random.choice([-1, 1], size=num_steps, p=[p, 1-p]))
print(positions)

# Plot the results
plt.plot(range(num_steps+1), np.concatenate(([0], positions)))
plt.xlabel('Time')
plt.ylabel('Position')
plt.title(f'1-D Random Walk with p = {p} (500 steps)')
plt.show()

# # Calculate the simulated probability
# print((positions[-1] == 0))
# print((positions[-1] == 0).mean())
# # simulated_probability = (positions[-1] == 0).mean()
# num=0
# for i in positions:
#     if(i==0):
#         num+=1
# simulated_probability=num/len(positions)
# print(f'Simulated Probability: {simulated_probability:.4f}')
```
